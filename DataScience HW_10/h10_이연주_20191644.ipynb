```
#Decision Tree Practice
#import modules
import requests
import random
from collections import Counter
```

```
#file open
iris_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
r = requests.get(iris_url)

with open('iris.data', 'wb') as f:
  f.write(r.content)
```

```
#iris데이터 불러오기

vectors = []
answers = []

with open('iris.data', 'r') as t:
  for line in t:
    if len(line.strip()) != 0:
      items = line.strip().split(",")
      vectors.append(tuple(float(i) for i in items[:4]))
      answers.append(items[4])
                
zipped = list(zip(vectors, answers)) #묶기

print("V : {}\nA : {}".format(vectors, answers))
print("Z : {}".format(zipped))
```

```
#Test/Training set 만들기
random.shuffle(zipped) #데이터 섞어서 80만 train 20은 test
train_size =  int(len(zipped) * 0.8)
train_x, train_y = zip(*zipped[:train_size]) #training data
test_x, test_y = zip(*zipped[train_size:]) #test data

print("train_X : {}, train_Y : {}".format(train_x, train_y))
print("test_X : {}, test_Y : {}".format(test_x, test_y))
```

```
#gini score 구하기
def gini_score(items):
  counter = Counter(items)
  #  print(len(items))
  #  print(sum(count.values()))
  total_count = sum(counter.values())
  gini = 1
  for item, count in counter.items():
    gini -= (count / total_count)**2

  return gini

gini_score(train_y)
```

```
#Numeric data에서 나뉘는 지점 찾기
def find_split_point_of_a_field(pairs):
  #pairs : (4.3, 'iris-~~'), (4.2, 'iris-~~') => pairs들의 list
  sort_pairs = sorted(pairs, key=lambda x: x[0])
  #print(sort_pairs)
  split_point, min_gini = 0, 99

  for i in range(1, len(sort_pairs)):

  #i-1 과 i가 다르면 둘 사이의 평균을 split_point로 두고 그 point의 gini가 가장 작은 값을 split point로 최종 선택
    if sort_pairs[i - 1][0] != sort_pairs[i][0]:
      g1 = gini_score(s[1] for s in sort_pairs[:i]) #구간의 giniScore 계산
      g2 = gini_score(s[1] for s in sort_pairs[i:])
      #gini = g1 * i / len(sort_pairs) + g2 * (len(sort_pairs))-i) / len(sort_pairs)

      pro = i / len(sort_pairs)
      gini = g1 * pro + g2 * (1 - pro)

      if min_gini > gini: 
        split_point, min_gini = (sort_pairs[i - 1][0] + sort_pairs[i][0]) / 2, gini
  return split_point, min_gini


# test 
zip_T = zip((x[0] for x in train_x), train_y)
print(find_split_point_of_a_field(zip_T))
```

```
#data를 실제로 split_point를 이용해서 나누쟈~~
def split_data(X, Y):
  num_fields = len(X[0])
  min_fid, min_sp, min_gini= -1, -1, 99 #gini값이 제일 작은 fid를 찾기 위한 변수

  #find least fid
  for fid in range(num_fields):
    sp, gini = find_split_point_of_a_field(zip((x[fid] for x in X), Y))
    if min_gini > gini:
      min_fid = fid
      min_sp = sp
      min_gini = gini
    #print(fid, sp, gini)
  #print(min_fid, min_sp, min_gini)

  node = {}
  node['sp'] = min_sp
  node['fid'] = min_fid
  node['gini'] = min_gini
  #print(min_fid, min_sp)
  #fid 보다 작은 부분들은 left, 크면 right로 해서 split해줌
  node['left'] = tuple(zip(*[(x, y) for x, y in zip(X, Y) if x[min_fid] < min_sp]))
  node['right']= tuple(zip(*[(x, y) for x, y in zip(X, Y) if x[min_fid] >= min_sp]))
  return node

split_data(train_x, train_y)
```

```
#재귀적으로 돌려서 tree를 만들장
#threshold차이가 많이나면 나누고, 그렇지 않으면 나눌필요 xx
def decision_tree(X, Y, threshhold):
  origin_gini = gini_score(Y)
  node = split_data(X, Y)
  if origin_gini - node['gini'] <= threshhold:#weight average
    # 나누지 않는다
    counter = Counter(Y)
    ans, c = counter.most_common(1)[0]
    return (ans, c/sum(counter.values()))
  else: #Data를 나눈다
    #XL, YL = node['left']
    node['left'] = decision_tree(*node['left'], threshhold)
    #XR, YR = node['right']
    node['right'] = decision_tree(*node['right'], threshhold)
    return node

#decision_tree(train_x, train_y, 0.1)
# decision_tree(train_x, train_y, 0)
# decision_tree(train_x, train_y, 0.005
```

```
#어디에 값이 속하는지 찾기
def predict(x , tree):
  # x = (2.3, 2.5, 6.3, 5.2)... 처럼 값이 들어옴
  if 'fid' not in tree: #종료조건 ~>  마지막값이 tuple같이 들어옴
    return tree

  if x[tree['fid']] < tree['sp']:
    #tree의 left
    return predict(x, tree['left'])
  else:
    return predict(x, tree['right'])

tree = decision_tree(train_x, train_y, 0)
print(predict(test_x[0], tree)) #실제
print(predict(test_x[0], test_y[0])) #예측값
```

```
#TRAIN_ACCURACY

train_accuracy = 0
for x, y in zip(train_x, train_y):
  if predict(x, tree)[0] == y:
    train_accuracy += 1

train_accuracy /= len(train_x)
print("Train Acc : ", train_accuracy) #1.0 나옴 ~> 다맞췄다

#TEST_ACCURACY

test_accuracy = 0
for x, y in zip(test_x, test_y):
  if predict(x, tree)[0] == y:
    test_accuracy += 1

test_accuracy /= len(test_x)
print("Test Acc : ", test_accuracy) #0.9 나옴 ~> threshhold 수정해봐도 좋을둣
```
